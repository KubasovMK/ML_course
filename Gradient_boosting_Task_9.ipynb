```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.metrics import f1_score
from sklearn.preprocessing import OrdinalEncoder, OneHotEncoder

# Загрузка данных
data = pd.read_csv('WA_Fn-UseC_-Telco-Customer-Churn.csv')

# Разделение данных на обучающую и тестовую выборки
train_data, test_data = train_test_split(data, test_size=0.2, stratify=data['Churn'], random_state=1)

# Удаление признака идентификатора клиента
train_data = train_data.drop('customerID', axis=1)
test_data = test_data.drop('customerID', axis=1)

# Замена неявных пропусков в числовых признаках на 0
numerical_cols = train_data.select_dtypes(include=['int64', 'float64']).columns
train_data[numerical_cols] = train_data[numerical_cols].fillna(0)
test_data[numerical_cols] = test_data[numerical_cols].fillna(0)

# Определение бинарных и категориальных признаков
binary_cols = [col for col in train_data.columns if train_data[col].nunique() == 2 and col != 'Churn']
categorical_cols = [col for col in train_data.select_dtypes(include=['object']).columns
                    if col not in binary_cols and col != 'Churn']

# Преобразование бинарных признаков через OrdinalEncoder
ordinal_encoder = OrdinalEncoder()
train_data[binary_cols] = ordinal_encoder.fit_transform(train_data[binary_cols])
test_data[binary_cols] = ordinal_encoder.transform(test_data[binary_cols])

# Преобразование категориальных признаков через OneHotEncoder
onehot_encoder = OneHotEncoder(sparse_output=False, handle_unknown='ignore')
train_encoded = onehot_encoder.fit_transform(train_data[categorical_cols])
test_encoded = onehot_encoder.transform(test_data[categorical_cols])

# Создание DataFrame для закодированных категориальных признаков
encoded_cols = onehot_encoder.get_feature_names_out(categorical_cols)
train_encoded_df = pd.DataFrame(train_encoded, columns=encoded_cols, index=train_data.index)
test_encoded_df = pd.DataFrame(test_encoded, columns=encoded_cols, index=test_data.index)

# Удаление исходных категориальных колонок
train_data = train_data.drop(categorical_cols, axis=1)
test_data = test_data.drop(categorical_cols, axis=1)

# Объединение всех признаков
X_train = pd.concat([train_data.drop('Churn', axis=1), train_encoded_df], axis=1)
X_test = pd.concat([test_data.drop('Churn', axis=1), test_encoded_df], axis=1)

# Преобразование целевой переменной в числовой формат
y_train = train_data['Churn'].map({'Yes': 1, 'No': 0})
y_test = test_data['Churn'].map({'Yes': 1, 'No': 0})

# Обучение Random Forest
rf = RandomForestClassifier(random_state=1)
rf.fit(X_train, y_train)
rf_train_f1 = f1_score(y_train, rf.predict(X_train), average='macro')
rf_test_f1 = f1_score(y_test, rf.predict(X_test), average='macro')

# Обучение Gradient Boosting
gb = GradientBoostingClassifier(random_state=1)
gb.fit(X_train, y_train)
gb_train_f1 = f1_score(y_train, gb.predict(X_train), average='macro')
gb_test_f1 = f1_score(y_test, gb.predict(X_test), average='macro')

# Разница в метрике на тесте
improvement = gb_test_f1 - rf_test_f1

print(f"Random Forest: F1 (train) = {rf_train_f1:.4f}, F1 (test) = {rf_test_f1:.4f}")
print(f"Gradient Boosting: F1 (train) = {gb_train_f1:.4f}, F1 (test) = {gb_test_f1:.4f}")
print(f"Улучшение F1-score на тесте (Gradient Boosting vs Random Forest): {improvement:.2f}")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.metrics import f1_score
from sklearn.preprocessing import OrdinalEncoder, OneHotEncoder

# Загрузка данных
data = pd.read_csv('WA_Fn-UseC_-Telco-Customer-Churn.csv')

# Разделение данных на обучающую и тестовую выборки
train_data, test_data = train_test_split(data, test_size=0.2, stratify=data['Churn'], random_state=1)

# Удаление признака идентификатора клиента
train_data = train_data.drop('customerID', axis=1)
test_data = test_data.drop('customerID', axis=1)

# Замена неявных пропусков в числовых признаках на 0
numerical_cols = train_data.select_dtypes(include=['int64', 'float64']).columns
train_data[numerical_cols] = train_data[numerical_cols].fillna(0)
test_data[numerical_cols] = test_data[numerical_cols].fillna(0)

# Определение бинарных и категориальных признаков
binary_cols = [col for col in train_data.columns if train_data[col].nunique() == 2 and col != 'Churn']
categorical_cols = [col for col in train_data.select_dtypes(include=['object']).columns
                    if col not in binary_cols and col != 'Churn']

# Преобразование бинарных признаков через OrdinalEncoder
ordinal_encoder = OrdinalEncoder()
train_data[binary_cols] = ordinal_encoder.fit_transform(train_data[binary_cols])
test_data[binary_cols] = ordinal_encoder.transform(test_data[binary_cols])

# Преобразование категориальных признаков через OneHotEncoder
onehot_encoder = OneHotEncoder(sparse_output=False, handle_unknown='ignore')
train_encoded = onehot_encoder.fit_transform(train_data[categorical_cols])
test_encoded = onehot_encoder.transform(test_data[categorical_cols])

# Создание DataFrame для закодированных категориальных признаков
encoded_cols = onehot_encoder.get_feature_names_out(categorical_cols)
train_encoded_df = pd.DataFrame(train_encoded, columns=encoded_cols, index=train_data.index)
test_encoded_df = pd.DataFrame(test_encoded, columns=encoded_cols, index=test_data.index)

# Удаление исходных категориальных колонок
train_data = train_data.drop(categorical_cols, axis=1)
test_data = test_data.drop(categorical_cols, axis=1)

# Объединение всех признаков
X_train = pd.concat([train_data.drop('Churn', axis=1), train_encoded_df], axis=1)
X_test = pd.concat([test_data.drop('Churn', axis=1), test_encoded_df], axis=1)

# Преобразование целевой переменной в числовой формат
y_train = train_data['Churn'].map({'Yes': 1, 'No': 0})
y_test = test_data['Churn'].map({'Yes': 1, 'No': 0})

# Исследование влияния max_depth на F1-score
max_depths = range(1, 21)
train_f1_scores = []
test_f1_scores = []

for depth in max_depths:
    gb = GradientBoostingClassifier(max_depth=depth, random_state=1)
    gb.fit(X_train, y_train)
    train_f1 = f1_score(y_train, gb.predict(X_train), average='macro')
    test_f1 = f1_score(y_test, gb.predict(X_test), average='macro')
    train_f1_scores.append(train_f1)
    test_f1_scores.append(test_f1)

# Визуализация
plt.figure(figsize=(10, 6))
plt.plot(max_depths, train_f1_scores, label='Train F1 (macro)', marker='o')
plt.plot(max_depths, test_f1_scores, label='Test F1 (macro)', marker='o')
plt.xlabel('max_depth')
plt.ylabel('F1-score (macro)')
plt.title('Влияние max_depth на качество Gradient Boosting')
plt.legend()
plt.grid()
plt.show()
```
